```
import pandas as pd                        
from pytrends.request import TrendReq
pytrend = TrendReq()
```

```
# https://towardsdatascience.com/google-trends-api-for-python-a84bc25db88f
```

```
pytrend.build_payload(kw_list=['python','r language','machine learning','artificial intelligence'])
# Interest by Region
df = pytrend.interest_by_region()
df.head(10)
```

```
df.reset_index().plot(x='geoName', y='python', figsize=(120, 10), kind ='bar')
```

```
df.reset_index().plot(x='geoName', y='r language', figsize=(120, 10), kind ='bar')
```

```
# Get Google Hot Trends data
df = pytrend.trending_searches(pn='brazil')
df.head()
```
